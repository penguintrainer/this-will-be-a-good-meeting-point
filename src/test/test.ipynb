```python
import numpy as np
import folium
from geopy.geocoders import Nominatim
```

```python
# 重心を計算したい地名一覧
areas = ["福岡県田川市","明治大学","北見工業大学","筑波大学","東京タワー","スカイツリー","富士山"]
```

```python
# 緯度経度の取得
points = []
for area in areas:
  geolocator = Nominatim(user_agent="user-id")
  location = geolocator.geocode(area)
  tmp_point = np.array([location.latitude, location.longitude])
  points.append(tmp_point)
#display(points)
```

```python
# 重心の計算
mean_latitudes = np.array(points)[:,0].mean()
mean_longitude = np.array(points)[:,1].mean()

# 描写に関する設定
folium_map = folium.Map(location=[mean_latitudes,mean_longitude],zoom_start=8)
folium.Marker(location=[mean_latitudes,mean_longitude],icon=folium.Icon(color="pink", icon="hand-down"),popup=folium.map.Popup("重心",max_width=10,show = True)).add_to(folium_map)
for i,point in enumerate(points):
  folium.Marker(location=point,popup=folium.map.Popup(areas[i])).add_to(folium_map)

display(folium_map)
```
